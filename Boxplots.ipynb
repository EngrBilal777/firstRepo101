```python
#how to remove error bars

import seaborn as sns
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from numpy import median
from numpy import mean
sns.set_style(style=None, rc=None)

## adding x limit and y limit
# load dataset
Boat = sns.load_dataset("titanic")
Boat.head()

# # draw a line plot
# sns.barplot(x="class", y= "fare", hue="sex", data=Boat, estimator=mean)
# plt.title("Titanic plot")

# plt.show()
```

```python
sns.boxplot(x="fare",
           y="sex",
           showmeans= True, meanprops = {"marker":"*", "markersize":"12", "markeredgecolor":"red"}, data=Boat)
plt.xlabel("How many survived"),
plt.ylabel("age(years)"),
plt.title("Box plot drowned")
```

```python
tips=sns.load_dataset("tips")
sns.FacetGrid(tips)
```

```python
sns.FacetGrid(tips, col="time", row="sex")
```

```python
g = sns.FacetGrid(tips, col="time",  row="sex")
g.map(sns.scatterplot, "total_bill", "tip")
```

```python
g = sns.FacetGrid(tips, col="time",  row="sex")
g.map_dataframe(sns.histplot, x="total_bill")
```

```python
g = sns.FacetGrid(tips, col="time", hue="sex")
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
g.add_legend()
```

```python
g = sns.FacetGrid(tips, col="day", height=3.5, aspect=.65)
g.map(sns.histplot, "total_bill")
```

```python
g = sns.FacetGrid(tips, col="time", margin_titles=True)
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
g.refline(y=tips["tip"].median())
```
