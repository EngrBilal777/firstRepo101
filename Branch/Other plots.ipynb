## Multifacetplots

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
boat=sns.load_dataset("titanic")
boat.head()
```

```python
dots=sns.load_dataset("dots")
p= sns.color_palette('rocket_r')
dots.head()
```

```python
sns.relplot(data=dots, x="time", y="firing_rate", hue= "coherence", 
           size="choice", col="align", kind="line", size_order=["T1", "T2"], palette=p, facet_kws=dict(sharex=False))
```
