### Tuples

##### 2- Enclosed in round braces () or paranthesis
##### 3- Different kind of elements can be stored
##### 4- Once elements are stored we cannot change them (unmutateable)

```python
tuple1=(1, "python", 1.5, True)
tuple1
```

```python
type(tuple1)
```
